### 🛒 **Günlük Hayattan Bir Örnek: İnternetten Sipariş Vermek**

Bir alışveriş sitesinden ürün alırken aslında arka planda çalışan karmaşık bir **algoritmanın** parçası oluruz:

1.  **Ürünü seç** ve "Sepete Ekle" butonuna bas.

### 🧮 **Uygulama: Gelecekteki Yaşınızı Hesaplayalım**

Şimdi öğrendiğimiz **"Girdi → İşlem → Çıktı"** modelini gerçek bir kod üzerinde görelim. Kullanıcıdan yaşını alacağız ve 5 yıl sonra kaç yaşında olacağını hesaplayacağız.

#### **🔑 Kodun İçindeki "Etiketler" (Değişkenler)**
Bilgisayarın aldığı bilgileri aklında tutması için onlara isimler veririz:
* `gelen_veri`: Kullanıcının klavyeden yazdığı ham bilgiyi (metni) tutan etikettir.
* `yas`: Bu metni matematiksel bir sayıya çevirdiğimizde kullandığımız yeni etikettir.

#### **⚙️ Dönüştürme Aracı: `int()`**
Python'da dışarıdan gelen her bilgi başlangıçta bir **"Metin"** (String) olarak kabul edilir. Metinlerle matematiksel toplama yapamazsınız.
* `int()` fonksiyonu: İçine yazılan metni, üzerinde işlem yapılabilecek bir **"Tam Sayı"** (Integer) haline getirir.

#### **📢 Ekrana Yazdırma Kuralları**
`print()` fonksiyonu içine bir şeyler yazarken iki farklı yöntem kullanırız:
1.  **Metin Yazdırmak:** Tırnak işareti içinde yazılır. Örn: `"Yaşınız:"`
2.  **Değişken Yazdırmak:** Tırnak **olmadan** doğrudan adı yazılır. Örn: `yas`
3.  **İkisini Birleştirmek:** Aralarına **virgül ( , )** koyarak ikisini aynı anda ekrana yazdırabiliriz.



> 🚀 **Hadi Deneyiniz:** Aşağıdaki kod hücresini çalıştırınız! `print` içindeki virgülün metin ile sayıyı nasıl ayırdığına dikkat ediniz.

````python
# 1. GİRDİ: Kullanıcıdan yaş bilgisini alıyoruz.
gelen_veri = input("Lütfen şu anki yaşınızı giriniz: ")

# 2. İŞLEM: Metni sayıya (integer) çeviriyoruz.
yas = int(gelen_veri)

# 3. ÇIKTI: Metin ve değişkeni virgül ile birleştirip yazdırıyoruz.
print("5 yıl sonra tam olarak şu yaşta olacaksınız:", yas + 5)
````

#### **🔍 Neler Oldu?**
Yazdığımız bu kısa kod aslında üç aşamalı ana modelimizi başarıyla çalıştırdı:

* 📥 **Girdi:** `input()` fonksiyonu ile yaş bilgisi kullanıcıdan alındı.
* ⚙️ **İşlem:** `int()` ile sayıya çevrildi ve üzerine **5** eklendi.
* 📤 **Çıktı:** `print()` ile sonuç anlamlı bir cümleyle ekrana yazdırıldı.




**Sıradaki Adım:** Bilgisayara kod yazmadan önce kendi dilimizde nasıl plan yapacağımızı (Sözde Kod) öğreneceğiz.

## 📝 **2.3. Algoritma Yazma: Kodlamadan Önce Planlama**

Bir problemi çözmeye çalışırken doğrudan kod yazmaya başlamak, haritasız yola çıkmaya benzer. Bunun yerine önce **Sözde Kod (Pseudocode)** kullanırız.

### 🤖 **Sözde Kod (Pseudocode) Nedir?**
Bilgisayarın teknik kurallarına takılmadan, adımları **kendi dilimizde** basitçe ifade etmektir.

**Örnek Problem:** Girilen sayı tek mi, çift mi?


#### 📋 **Sözde Kod Örneği: Tek/Çift Kontrolü**

Bir yazılımcı gibi düşünerek planımızı şu şekilde kağıda döküyoruz:

```text
BAŞLA
    OKU sayı
    EĞER (sayı % 2 == 0) İSE       # Not: "EĞER (sayı MOD 2 == 0) İSE" de denebilir.
        YAZ "Sayı Çifttir"
    DEĞİLSE
        YAZ "Sayı Tektir"
    EĞER_BİTTİ
BİTTİ

### 🧮 **Python Uygulaması: Tek mi, Çift mi?**

Planladığımız sözde kodu (pseudocode) şimdi gerçek Python koduna dönüştürüyoruz.

Burada Python'un en güçlü özelliklerinden biri olan **"Karar Yapıları"** ile tanışıyoruz.

#### **🛠️ Yeni Araçlarımız:**
* **`if` (Eğer):** Bir koşulun doğru olup olmadığını kontrol eder. Sonuna mutlaka **iki nokta ( : )** konur.
* **`else` (Değilse):** Koşul doğru değilse yapılacak olan "B planı"dır.
* **`% 2 == 0`:** Sayının 2'ye bölümünden kalanın 0'a **eşit olup olmadığını** kontrol eder. (`==` işareti "eşit mi?" diye sormak için kullanılır).



> 🚀 **Hadi Deneyiniz:** Aşağıdaki kod hücresini çalıştırınız. Farklı sayılar (örneğin 15, 22, 100) girerek Python'un nasıl karar verdiğini gözlemleyiniz!

````python
# 1. GİRDİ: Kullanıcıdan kontrol edilecek sayıyı alıyoruz
sayi = int(input("Lütfen bir sayı giriniz: "))

# 2. İŞLEM ve KARAR: Sayının 2'ye bölümünden kalanı kontrol ediyoruz
if sayi % 2 == 0:
    # Koşul doğruysa (Kalan 0 ise) burası çalışır
    print("Girdiğiniz sayı: Çifttir")
else:
    # Koşul yanlışsa (Kalan 0 değilse) burası çalışır
    print("Girdiğiniz sayı: Tektir")
````

```test
Sözde Kod (Pseudocode)  | Python Karşılığı | Açıklama
------------------------|------------------|-----------------------------------------
OKU                     | input()          | Kullanıcıdan veri alır.
YAZ                     | print()          | Ekrana çıktı verir.
MOD                     | %                | Bölme işleminden kalanı verir.
EĞER / İSE              | if / :           | Koşullu dallanma başlatır.
DEĞİLSE                 | else             | Koşul dışındaki tüm durumlar.

```text
⚠️ Önemli Hatırlatma: input() ile alınan veriler varsayılan olarak "Metin" tipindedir.
Matematiksel işlem (MOD, Toplama vb.) yapabilmek için bu veriyi int() fonksiyonu ile Tam Sayı'ya çevirmemiz gerekir.

# Kullanıcıdan alınan veriyi anında tam sayıya çeviriyoruz.
sayi = int(input("Bir sayı giriniz: "))

# Artık matematiksel işlem yapabiliriz.
sonuc = sayi % 2
print("Sayının 2'ye bölümünden kalan:", sonuc)

## **2.5. Adım Adım Problem Çözme**

### **Örnek Problem-2: Ortalama Hesaplama**
*Soru: Kullanıcıdan 3 not al, ortalama 50 veya daha büyükse "GEÇTİ", değilse "KALDI" yaz.*


#### **2.5.1. Algoritma (Doğal Dil ile)**
1. **BAŞLA**
2. Kullanıcıdan birinci, ikinci ve üçüncü notları iste.
3. Üç notu toplayıp üçe bölerek ortalamayı bul.
4. Eğer ortalama 50'ye eşit veya büyükse ekrana "GEÇTİ" yazdır.
5. Eğer ortalama 50'den küçükse ekrana "KALDI" yazdır.
6. **BİTİR**


#### **2.5.2. Sözde Kod (Pseudo Code)**
```text
BAŞLA
  OKU n1, n2, n3
  ort = (n1 + n2 + n3) / 3
  EĞER ort >= 50 İSE
    YAZ "GEÇTİ"
  DEĞİLSE
    YAZ "KALDI"
  EĞER_BİTTİ
BİTİR

#### **2.5.4. Python Kodu**

Bu örnekte kullanıcıdan alınan 3 notun ortalaması hesaplanır ve mantıksal bir karşılaştırma yapılarak sonuç ekrana yazdırılır.

**Programın Mantığı:**
1. **Veri Girişi (OKU):** `input()` ile alınan veriler `int()` ile tam sayıya dönüştürülür.
2. **Hesaplama (İŞLEM):** Matematiksel operatörler kullanılarak ortalama bulunur.
3. **Karar Yapısı (KARAR / YAZ):** `if-else` bloğu ile ortalama **50** sınırı kontrol edilir.

---

````python
# 1. Adım: Verileri al ve sayıya çevir (OKU)
n1 = int(input("1. Notu giriniz: "))
n2 = int(input("2. Notu giriniz: "))
n3 = int(input("3. Notu giriniz: "))

# 2. Adım: Hesaplama yap (İŞLEM)
ort = (n1 + n2 + n3) / 3

# 3. Adım: Koşulu kontrol et ve yazdır (KARAR / YAZ)
if ort >= 50:
    print(f"Ortalamanız: {ort:.2f} - GEÇTİ ✅")
else:
    print(f"Ortalamanız: {ort:.2f} - KALDI ❌")
````
