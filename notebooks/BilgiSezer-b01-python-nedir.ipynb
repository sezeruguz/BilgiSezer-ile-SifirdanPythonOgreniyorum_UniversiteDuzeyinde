## **1.1. Programlama Dili Nedir?**

Bilgisayarlar bizim dilimizi anlamaz.
Sadece elektrik sinyallerini anlar (0 ve 1).

Bu nedenle insanlar bilgisayara komut verebilmek için
**programlama dilleri** geliştirmiştir.

Programlama dili; tıpkı bir şefe ne yapacağını söyleyen bir yemek tarifi veya yabancı dil bilmeyen iki kişi arasındaki tercüman gibidir.

---
## **1.2. Python Nedir?**

Python, okunması kolay olacak şekilde tasarlanmış
yüksek seviyeli bir programlama dilidir.

Ama daha önemli tanım olarak şöyle söyleyebiliriz:

> Python, problemi çözmeye odaklanmamızı sağlayan bir dildir.

## **1.3. Neden Python Öğreniyoruz?**

Çünkü Python:

- Kolay okunur.
- Hataları anlaşılırdır.
- Birçok alanda kullanılabilir.

## **1.4. Python Nerelerde Kullanılır?**

- Yapay Zeka
- Veri Analizi
- Web Geliştirme
- Otomasyon
- Bilimsel Hesaplama
- Robotik

---
## **1.5. Python Nasıl Çalışır?**

Diğer bazı dillerde program önce tamamen makine diline çevrilir.

Python'da ise:

"Kod → Python yorumlayıcı → Çalıştırma"

Bu nedenle Python'a **yorumlanan dil** denir.

---
## **1.6. Derlenen ve Yorumlanan Diller**

Derlenen diller:
Program tamamen çevrilir → sonra çalışır

Yorumlanan diller:
Program satır satır çalışır.

Python ikinci gruptadır.
